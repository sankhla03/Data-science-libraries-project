#iteration over Array


```python
a = np.arange(0,45,5)
a = a.reshape(3,3)
a
```

```python
for x in np.nditer(a):
  print(x)
```

```python
# iteration order (c- style and y-stlye)
```

```python
print(a)
for x in np.nditer(a, order = 'C'):
  print(x)

for x in np.nditer(a, order = 'F'):
                   print(x)
```

```python
a = np.array([[1,2],[3,4]])
print("first array")
print(a)
b = np.array([[5,6],[7,8]])
print("second array")
print(b)
print('\n')
print("joining the two arrays along axis 0:" )
print(np.concatenate((a,b)))
print('\n')
print("joining the two arrays along axis 1:" )
print(np.concatenate((a,b), axis =1))
print('\n')
```

```python
#find the total number and locations of missing value in the array
z = np.random.rand(10,10)
z[np.random.randint(10,size =5), np.random.randint(10,size =5)] = np.nan
z
```

```python
print("Total number of missing values: \n", np.isnan(z).sum())
```

```python
print("Index of missing values: \n", np.argwhere(np.isnan(z)))
print('\n')
inds = np.where(np.isnan(z))
print(inds)
print('\n')
z[inds] =0
print(z)
```
